```python
import numpy as np
from skimage import io
from matplotlib import pyplot as plt
img = io.imread('/content/BSE_Image.jpg')
plt.imshow(img)
img.shape
```

```python
img2 = img.reshape(-1,3)
img2.shape
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4, random_state=0, n_init="auto").fit(img2)


"""
kmeans.predict([[0, 0], [12, 3]])
kmeans.cluster_centers_
"""
```

```python
kmeans.cluster_centers_

# 4x3 matrix
# 4 centers in 3 channels
# the value is intensity
```

```python
kmeans.labels_.flatten()
```

```python
a = img.shape
b = np.array(a)
c = b[:-1]

res = kmeans.labels_.reshape(c)
plt.subplot(121)
plt.imshow(res)
plt.subplot(122)
plt.imshow(img)
```
