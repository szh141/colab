```python
import matplotlib.pyplot as plt
import numpy as np

x,y = np.arange(-50,50),np.arange(-50,50)
X,Y = np.meshgrid(x,y)
circle = np.sqrt(X**2 + Y**2)

plt.subplot(121), plt.imshow(circle)
plt.subplot(122), plt.contourf(X,Y,circle)
plt.show()
```

plt.contour (group the intensity into heights) vs plt.imshow (smoother transition in intensity)

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

x = np.linspace(1, 4, 20)
y = np.linspace(1, 4, 20)
X, Y = np.meshgrid(x, y)
Z = np.sqrt((X - 2.5)**2 + (Y - 2.5)**2)

plt.figure(figsize=(8, 6))

plt.subplot(2, 3, 1)
plt.ylabel('y')
plt.grid()
plt.title('cm.jet')
plt.contourf(X, Y, Z, cmap=cm.jet)

plt.subplot(2, 3, 2)
plt.grid()
plt.title('cm.gray')
plt.contourf(X, Y, Z, cmap=cm.gray)

plt.subplot(2, 3, 3)
plt.grid()
plt.title('cm.viridis')
plt.contourf(X, Y, Z, cmap=cm.viridis)

plt.subplot(2, 3, 4)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.title('cm.winter')
plt.contourf(X, Y, Z, cmap=cm.winter)

plt.subplot(2, 3, 5)
plt.xlabel('x')
plt.grid()
plt.title('cm.seismic')
plt.contourf(X, Y, Z, cmap=cm.seismic)

plt.subplot(2, 3, 6)
plt.xlabel('x')
plt.grid()
plt.title('cm.Set3')
plt.contourf(X, Y, Z, cmap=cm.Set3)

plt.show()
```
